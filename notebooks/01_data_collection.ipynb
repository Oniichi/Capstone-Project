```python
import yfinance as yf
import pandas as pd

tickers = {
    "gold": "GC=F",
    "dxy": "DX-Y.NYB",
    "vix": "^VIX",
    "spy": "SPY",
    "oil": "CL=F",
    "treasury_10y": "^TNX",
    "eurusd": "EURUSD=X"
}

start_date = "2003-01-01"
end_date = "2024-12-31"

raw_data = {}

for name, ticker in tickers.items():
    print(f"Downloading {name} ({ticker})...")
    data = yf.download(ticker, start=start_date, end=end_date)
    raw_data[name] = data
    data.to_csv(f"../data/raw/{name}.csv")

raw_data["gold"].head()
```
