```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

# Load dataset
df = pd.read_csv("/Users/erion/Desktop/MSCF2/ADA/Capstone-Project/data/processed/merged_dataset.csv",
                 parse_dates=["Date"],
                 index_col="Date")

df.head()
```

```python
plt.figure(figsize=(12,4))
df['gold'].plot(title="Gold Daily Returns (20032024)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['gold'], bins=100, kde=True)
plt.title("Distribution of Gold Returns")
plt.show()
```

```python
adf_result = adfuller(df['gold'])
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
```

```python
plt.figure(figsize=(12,4))
df['gold'].rolling(30).std().plot()
plt.title("30-Day Rolling Volatility of Gold Returns")
plt.show()
```

```python
df.plot(subplots=True, figsize=(12,10), title="All Variables Over Time")
plt.tight_layout()
plt.show()
```
